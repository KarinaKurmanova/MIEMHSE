```python
import numpy as np

from math import log

def  forward_differences(f, x, h):
    df = -1.5*f(x)+2.0*f(x + h) - 0.5*f(x +2.0*h)
    return df / (h)
```

```python
f_xlnx=lambda x: x*x*log(x)
```

```python
forward_differences(f_xlnx, 0.7, 1/200) 
```

```python

x = np.linspace(1e-5, 1e-4, 101)
```

```python
y=[]
for i in x:
    y.append(forward_differences(f_xlnx, i ,1e-4 ) )
    
    
```

```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
plt.plot(x, y, 'o', label='Abs error')

plt.legend(loc='best')
```

При х->0 значение производной тоже стремится к нулю.


```python
xnew = np.linspace(0.1, 3, 200)
```

```python
ynew=[]
for i in xnew:
    ynew.append(forward_differences(f_xlnx, i ,1e-4 ) )
    
```

```python
plt.plot(xnew, ynew, 'o', label='Abs error')

plt.legend(loc='best')
```
